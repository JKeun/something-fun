## Plot 이미지 제작소

```python
from matplotlib.patches import Polygon

xx = np.linspace(-4, 4, 100)
rv = sp.stats.norm(0, 1)

fig = plt.figure(figsize=(10, 5))
y_labels = []

left_list = [0, -1, -1.5, -2]
right_list = [1.5, 1, -0.5, 2]
i_list = [1, 2, 3, 4]

with plt.style.context(('seaborn-white')):
    
    for i, left, right in zip(i_list, left_list, right_list):
        ax = fig.add_subplot(2,2,i)
        plt.plot(xx, rv.pdf(xx))
        plt.yticks(y_labels)
        plt.xlabel('Z')
        plt.ylabel("probability")

        ix = np.linspace(left, right)
        iy = rv.pdf(ix)
        verts = [(left, 0)] + list(zip(ix, iy)) + [(right, 0)]
        poly = Polygon(verts, edgecolor="#599CBB", facecolor="#ACCDDD")
        ax.add_patch(poly)

plt.savefig('./images/standard-narmal-dist1.png', bbox_inches='tight')
```

## Customized Matplotlib

```python
x = np.linspace(0, 20, 100)
with plt.style.context(('seaborn-white')):
    fig, axes = plt.subplots(nrows=2)
    for i in range(5):
        axes[0].plot(x, i * (x - 10)**2, '-o')
    for i in range(5):
        axes[1].plot(x, i * np.cos(x), '-')
    plt.show()
```

```python
data = np.random.normal(size=10000).reshape(-1, 2)
data[:, 1] = data[:, 1] * 0.5 + 1
with plt.style.context(('seaborn-white')):
    _ = plt.hist(data, bins=30, edgecolor='#FFFFFF')
```
