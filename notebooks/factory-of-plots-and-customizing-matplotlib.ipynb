### 이미지 제작소

```python
from matplotlib.patches import Polygon

xx = np.linspace(-4, 4, 100)
rv = sp.stats.norm(0, 1)

fig = plt.figure(figsize=(10, 7))
y_labels = []

left_list = [0, -1, -1.5, -2]
right_list = [1.5, 1, -0.5, 2]
i_list = [1, 2, 3, 4]

with plt.style.context(('seaborn-white')):
    
    for i, left, right in zip(i_list, left_list, right_list):
        ax = fig.add_subplot(2,2,i)
        plt.plot(xx, rv.pdf(xx))
        plt.yticks(y_labels)
        plt.xlabel('Z')
        plt.ylabel("probability")

        if i == 4:
            ix_left = np.linspace(-4, left)
            ix_right = np.linspace(right, 4)
            iy_left = rv.pdf(ix_left)
            iy_right = rv.pdf(ix_right)
            verts = [(-4, 0)] + list(zip(ix_left, iy_left)) + [(left, 0),
                    (right, 0)] + list(zip(ix_right, iy_right)) + [(4, 0)]
            
        else:
            ix = np.linspace(left, right)
            iy = rv.pdf(ix)
            verts = [(left, 0)] + list(zip(ix, iy)) + [(right, 0)]

        poly = Polygon(verts, edgecolor="#599CBB", facecolor="#ACCDDD")
        ax.add_patch(poly)
        
plt.savefig('./images/standard-narmal-dist1.png', bbox_inches='tight')
```

```python
from matplotlib.patches import Polygon

fig = plt.figure(figsize=(9, 7))
y_labels = []


mu_list = [100, 0]
std_list = [10, 1]
i_list = [1, 2]

with plt.style.context(('seaborn-white')):
    
    for i, mu, std in zip(i_list, mu_list, std_list):
        xx = np.linspace(mu - 5*std, mu + 5*std, 100)
        rv = sp.stats.norm(mu, std)
        
        ax = fig.add_subplot(2,1,i)
        plt.plot(xx, rv.pdf(xx))
        plt.yticks(y_labels)
        if i == 1:
            plt.xlabel("X")
        else:
            plt.xlabel("Z")
        plt.ylabel("probability")

        ix = np.linspace(mu, mu+std)
        iy = rv.pdf(ix)
        verts = [(mu, 0)] + list(zip(ix, iy)) + [(mu+std, 0)]
        poly = Polygon(verts, edgecolor="#599CBB", facecolor="#ACCDDD")
        ax.add_patch(poly)
        
plt.savefig('./images/standard-narmal-dist2.png', bbox_inches='tight')
```

```python
from matplotlib.patches import Polygon

fig = plt.figure(figsize=(9, 4))
y_labels = []


mu_list = [0]
std_list = [1]
i_list = [1]

with plt.style.context(('seaborn-white')):
    
    for i, mu, std in zip(i_list, mu_list, std_list):
        xx = np.linspace(mu - 5*std, mu + 5*std, 100)
        rv = sp.stats.norm(mu, std)
        
        ax = fig.add_subplot(1,1,i)
        plt.plot(xx, rv.pdf(xx))
        plt.yticks(y_labels)
        plt.xlabel("Z")
        plt.ylabel("probability")

        ix = np.linspace(mu+2*std, mu+5*std)
        iy = rv.pdf(ix)
        verts = [(mu+2*std, 0)] + list(zip(ix, iy)) + [(mu+5*std, 0)]
        poly = Polygon(verts, edgecolor="#599CBB", facecolor="#ACCDDD")
        ax.add_patch(poly)
        
plt.savefig('./images/standard-narmal-dist3.png', bbox_inches='tight')
```

## Customized Matplotlib

```python
x = np.linspace(0, 20, 100)
with plt.style.context(('seaborn-white')):
    fig, axes = plt.subplots(nrows=2)
    for i in range(5):
        axes[0].plot(x, i * (x - 10)**2, '-o')
    for i in range(5):
        axes[1].plot(x, i * np.cos(x), '-')
    plt.show()
```
